To make a deeper comparison, lets pick three differenct countries on three different continents with hugely different GDPs. When looking at the normalized energy consumption of each country, we can examine theri national energy mix. Morocco has a high gas usage with around 35% of the energy mix over the years, while at the same time fossil fuels are also used at a rate of around 30%. Coal consumption in Morocco has declined in the past decades, while oil consumption has increased, now making up of around 10% of the energy mix. 

```python
x.plot_consumption("Morocco",normalize=True)
```

France has a wildly different energy mix. Even though their gas consumption is about equally as high as Moroccos, their energy mix has changed after the 80s. Fossil fuels and coal have declined heavily, now being at around 17% and 11% respectively. Further, France has stepped up in nuclear power, consuming around 12-13% of it in their energy mix. Low carbon consumption has also increased in the recent years, sitting at around 13%

```python
x.plot_consumption("France",normalize=True)
```

China has and still does rely on gas, coal and fossil fuels. Only recently renewable energies have started to gain steam in China, which was and still continues to suffer from smog. However, with the trend going towards renewable energies, one can imagine that in ten years they will play a major role in the Chinese energy mix.

```python
x.plot_consumption("China",normalize=True)
```

The GDP of the three countries also tells an interesting story. China's GDP has been growing rapidly, even more so after Deng Xiaoping's opening of the country in 1990. France has been growing slowly but steadily, with Morocco having the slowest growth rate in the past decades.
