To make a deeper comparison, lets pick three differenct countries on three different continents with hugely different GDPs. For Morocco, their energy consumption is incredibly low when comparing to France and China. And even though China has more than 20 times the population of France, their energy consumption is just 5-6 fold of France's. In relation to each country's GDP, France and China produce a proportionate amount of emissions, while Morocco produces much more emissions in relation to its GDP. This could point towards inefficient technologies being used in production, lack of regulation or high-emission transport. 

```python
x.energy_compare(["Morocco","France","China"])
```

When looking at the normalized energy consumption of each country, we can examine their national energy mix. Morocco has a high oil usage with around 70% of the energy mix over the years, while at the same time fossil fuels are also used at a rate of around 15-25%. Coal and oil consumption in Morocco has declined in the past years, while gas and wind and solar have seen a premiere since 2010. 
Moroccos energy consumption has increased steadily with the current usage being around 5-fold what was being consumed in 1980.

```python
x.plot_consumption("Morocco",normalize=True)
x.plot_consumption("Morocco",normalize=False)
```

France has a wildly different energy mix. Most currently France heavily relies on oil and nuclear energy, both with 30% and 35% respectively. Nuclear has seen a high incline since the 1980s in France, while coal has been steadily declining for decades. Renewable energies are not playing a huge role in the French energy mix, but they have been visible with low percentages since around 2005. 
In comparison to Morocco and China, France is the only country where the total energy consumption has declined since 2008. The cause for this phenomenon could be the use of new, more energy efficient technologies or the introduction of a carbon tax by the EU. 

```python
x.plot_consumption("France",normalize=True)
x.plot_consumption("France",normalize=False)
```

China has and still does rely on fossil fuels. The coal usage of China has decreased relatively, when looking at the normalized values, but absolutely it has increased as China has developed its economy. Other important energy sources for China are oil, biofuel and gas. Even though China is one of the countries that invests the most in renewable technologies, it still has a long way to go before renewable technologies make up a significant part of their energy consumption.
In the bottom chart you can see the incredible growth that China went trough in the past decades. The country now uses around 8 times as much energy as in 1980, which is first caused by widespread industrialization, but also electrification and poverty alleviation in the whole country. 

```python
x.plot_consumption("China",normalize=True)
x.plot_consumption("China",normalize=False)
```

The GDP of the three countries also tells an interesting story. China's GDP has been growing rapidly, even more so after Deng Xiaoping's opening of the country in 1990. France has been growing slowly but steadily, with Morocco having the slowest growth rate in the past decades. 
China's GDP growth mirrors the growth of their energy consumption, but Morocco does not tell the same story. Their GDP did not grow in proportion to their energy consumption, as it has been stagnating for the past couple of decades.
